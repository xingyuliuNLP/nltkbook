#### display information about each text


```python
# average sentence length and lexical diversity appear to be characteristics of particular authors
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid)) # counts space characters without any linguistic processing
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid)) # includes punctuation
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(round(num_chars/num_words), # average word length
          round(num_words/num_sents), # average sentence length
          round(num_words/num_vocab), # lexical diversity score
          fileid)
```

```python
# conditional frequency distribution 
# text genre prediction
cfd = nltk.ConditionalFreqDist(
           (genre, word)
           for genre in brown.categories()
           for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

#### generating random text with bigrams

```python
# The bigrams() function takes a list of words and builds a list of consecutive word pairs. 
# in order to see the result and not a cryptic "generator object", we need to use the list() function
def generate_model(cfdist, word, num=15):
    for i in range(num):
        print(word, end=' ')
        word = cfdist[word].max() # most frequent bigram of living

text = nltk.corpus.genesis.words('english-kjv.txt')
bigrams = nltk.bigrams(text)
cfd = nltk.ConditionalFreqDist(bigrams)
cfd['living']
```

```python
generate_model(cfd, 'living')
```

```python
"""
compute what fraction of words in a text are not in the stopwords list
"""
def content_fraction(text):
     stopwords = nltk.corpus.stopwords.words('english')
     content = [w for w in text if w.lower() not in stopwords]
     return len(content) / len(text)

content_fraction(nltk.corpus.reuters.words())
```

#### lexical relations - hypernym, hyponym, meronym,holonym

```python
# from items to their components meronym
wn.synset('tree.n.01').part_meronyms()
```

```python
# from items to the things they are contained in (holonyms)
wn.synset('tree.n.01').member_holonyms()
```

```python
# The substance a tree is made of includes heartwood and sapwood
wn.synset('tree.n.01').substance_meronyms()
```

```python
# Some lexical relationships hold between lemmas, e.g., antonymy:
wn.lemma('supply.n.02.supply').antonyms()
```

```python
# 14
def supergloss(s):
    defS = s.name() + ': ' + s.definition() + '\n'
    for synset in s.hypernyms():
        defHyperS = synset.name() + ': ' + synset.definition() + '\n'
    for synset in s.hyponyms():
        defHypoS = synset.name() + ': ' + synset.definition()+ '\n'
    return defS, defHyperS, defHypoS
supergloss(computer)
```

```python
# 14
def supergloss(s):
    defis = ''
    defis = defis + s.name() + ': ' + s.definition() + '\n'
    for synset in s.hypernyms():
        defis = defis + synset.name() + ': ' + synset.definition() + '\n'
    for synset in s.hyponyms():
        defis = defis + synset.name() + ': ' + synset.definition() + '\n'
    return defis
print(supergloss(computer))
```

```python
# 18
def find_50_most_frequent_bigrams(text):
    bigram = list(nltk.bigrams(text))
    fdist = FreqDist(b for b in bigram if b[0].isalpha() and b[1].isalpha()
                                      and b[0] not in stopwords.words('english') 
                                      and b[1] not in stopwords.words('english')) 
    return fdist.most_common(50)

find_50_most_frequent_bigrams(text1)
```
